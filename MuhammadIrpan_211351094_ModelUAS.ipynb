```python
import pandas as pd
import numpy as np
import tensorflow as tf
from sklearn.preprocessing import StandardScaler
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Dense
from sklearn.model_selection import train_test_split
from sklearn.metrics import accuracy_score
```

```python
df = pd.read_csv("aids-virus-infection-prediction/AIDS_Classification.csv")
```

```python
df.head()
```

```python
df.info()
```

```python
df.shape
```

```python
df.drop(columns=['hemo','homo','drugs','z30','str2','treat','treat','offtrt','oprior','trt','strat','gender','symptom'], inplace=True)
```

```python
df.isnull().sum()
```

```python
df.describe().transpose()
```

```python
X = df.drop(columns='infected')
Y = df['infected']
```

```python
X_train, X_test, Y_train, Y_test = train_test_split(X, Y, test_size = 0.25)
```

```python
scaler = StandardScaler()
X_train = scaler.fit_transform(X_train)
X_test = scaler.transform(X_test)
```

```python
model = Sequential([
    Dense(64, activation='relu', input_shape=(X_train.shape[1],)),
    Dense(32, activation='relu'),
    Dense(1, activation='sigmoid')
])
```

```python
model.compile(optimizer='adam', loss='binary_crossentropy', metrics=['accuracy'])
```

```python
model.summary()
```

```python
model.compile(optimizer='adam', loss='binary_crossentropy', metrics=['accuracy'])
```

```python
result = model.fit(X_train, Y_train, epochs=20, batch_size=10)
```

```python
y_pred =model.predict(X_test)
y_pred = (y_pred>0.5)
```

```python
accuracy = accuracy_score(Y_test,y_pred)
print('Akurasi:', accuracy)
```

```python
model.save('./iris.h5')
```

```python
model = tf.keras.models.load_model('./iris.h5')
converter = tf.lite.TFLiteConverter.from_keras_model(model)
converter.optimizations = [tf.lite.Optimize.DEFAULT]
tflite_model = converter.convert()
open("./iris.tflite", "wb").write(tflite_model)
```
